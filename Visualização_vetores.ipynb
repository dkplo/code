```
rsa_vetores = list(rsa[:,0])
elgamal_vetores = list(elgamal[:,0])
ec_vetores = list(ec[:,0])
```

```
from sklearn import preprocessing
```

```
rsa_vetores_norm = preprocessing.normalize(rsa_vetores, norm='max')
elgamal_vetores_norm = preprocessing.normalize(elgamal_vetores, norm='max')
ec_vetores_norm = preprocessing.normalize(ec_vetores, norm='max')
```

```
for i in range(5):
    fig, axs = plt.subplots(1, 3, sharex=True, figsize=(15,7))
    axs[0].bar(x, rsa_vetores_norm[i])
    axs[0].set_title("RSA")

    # marker from TeX
    axs[1].bar(x, elgamal_vetores_norm[i])
    axs[1].set_title("Elgamal")

    # marker from path
    axs[2].bar(x, ec_vetores_norm[i])
    axs[2].set_title("EC")
    plt.show()
```

```
for i in range(5):
    df = pd.DataFrame({
        'rsa': rsa_vetores_norm[i],
        'elgamal': elgamal_vetores_norm[i],
        'ec': ec_vetores_norm[i],
    })
    ax = df.plot.kde(figsize=(15,7))
    plt.show()
```
